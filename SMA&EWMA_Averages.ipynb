```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
data = pd.read_csv("/content/drive/My Drive/TimeSeriesDataAnalysis/Data/airline_passengers.csv", index_col="Month", parse_dates=True)
data.head()
```

**SimpleMovingAverage SMA 6 & 12 Month**

```
data['6-month-sma'] = data['Thousands of Passengers'].rolling(window=6).mean()
```

```
data.head(10)
```

```
data['12-month-sma'] = data['Thousands of Passengers'].rolling(window=12).mean()
```

```
data.head(15)
```

```
data.plot(figsize=(16, 10), title="Passengers Simple Moving Average")
```

**ExponentiallyWeightedMovingAverage EWMA 6 & 12 Month**

```
data['ewma-12'] = data['Thousands of Passengers'].ewm(span=12).mean()
```

```
data[['Thousands of Passengers', 'ewma-12']].plot(figsize=(16, 10), title="Passengers EWMA 12M Average")
```
