```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import statsmodels.api as sm

from statsmodels.tsa.stattools import acovf, acf, pacf, pacf_yw, pacf_ols
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

from pandas.plotting import lag_plot
```

```
data_1 = pd.read_csv("/content/drive/My Drive/TimeSeriesDataAnalysis/Data/airline_passengers.csv", index_col="Month", parse_dates=True)
data_1.index.freq = 'MS' # monhly start data non stationary
data_1.index
```

```
data_2 = pd.read_csv("/content/drive/My Drive/TimeSeriesDataAnalysis/Data/DailyTotalFemaleBirths.csv", index_col="Date", parse_dates=True)
data_2.index.freq = 'D' # daily data stationary
data_2.index
```

```
data_1.head()
```

```
data_2.head()
```

**AutoCorrelation**

```
plt.figure(figsize=(16,10))
lag_plot(data_1['Thousands of Passengers'])
```

```
plt.figure(figsize=(16,10))
lag_plot(data_2['Births'])
```

```
data_1.plot(figsize=(16,10))
```

```
# acf plot
fig, ax = plt.subplots(figsize=(16,8))
plot_acf(data_1, lags=40,ax=ax, title='Thousand Of passengers ACF');
```

```
# acf plot
fig, ax = plt.subplots(figsize=(16,8))
plot_acf(data_2, lags=40,ax=ax, title='Daily Female Birth ACF');
```

```
# pacf plot
fig, ax = plt.subplots(figsize=(16,8))
plot_pacf(data_2, lags=40,ax=ax, title='Daily Female Birth PACF');
```
